```python
def getDetails(url):
    page_html = urllib2.urlopen(URL)
    page_beauty = BeautifulSoup(page_html, "html.parser")
#     Product Title
    dataTitle = page_beauty.find('span', attrs={'class': '_35KyD6'})
    title = dataTitle.text.strip()
    print('\033[1m' + title)
    
#     Product Price
    dataPrice = page_beauty.find('div', attrs={'class': '_1vC4OE _3qQ9m1'})
    price = dataPrice.text.strip()
    print("Price:"+" "+'\033[1m' + price)
    
#     Product Specifications
    dataHighlight = page_beauty.find('div', attrs={'class': '_3WHvuP'})
    print('\033[1m' + "Specification")

    # price = dataPrice.text.strip()
    data = dataHighlight.find_all('li')

    for d in data:
        nave = d.text.strip()
        print '\033[0m'+">>" + " " + nave
    print("-------------------------")
    dataSepc = page_beauty.find('div', attrs={'class': '_36fS1V'})
#     dta = dataSepc.str()
#     print type(dta)
    if dataSepc is None:
        print('Product Description is not available')
        print('-------------------------')
    else:
        print("\033[1m"+"Product Description:"+'\n'+"\033[0m"+dataSepc.text.strip())
        print("-------------------------")
#         specificationData = page_beauty.find_all('div', attrs={'class': '_3Rrcbo'})
#         for sub_data in specificationData:
#             head_main = page_beauty.find('div', attrs={'class': '_2lzn0o'})
#             print("\033[1m"+head_main.text.strip()+"\033[0m")
#     print("-------------------------")
    fullData = []
    subData = page_beauty.find_all('li', attrs={'class': '_3_6Uyw row'})
    for sp in subData:
        subtitle = sp.find_all('div', attrs={'class': '_3-wDH3 col col-3-12'})
        for st in subtitle:
            title = st.text.strip()
            fullData.append(title)
        subdesc = sp.find_all('li', attrs={'class': '_3YhLQA'})
        for sp in subdesc:
            desc = sp.text.strip()
            fullData.append(desc)

    b = zip(fullData[::2], fullData[1::2])
    for bbs in b:
        str1 = ':'.join(bbs)
        print str1
```
